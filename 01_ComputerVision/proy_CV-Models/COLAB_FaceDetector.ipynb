# Librerías

Importamos las librerías necesarias para poder reproducir el código

```python
# import dependencies

# from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Procesamiento

Importamos una imagen desde nuestro PC y la cargamos.

```python
img_dir = "./imagenes/face.jpg"
img = cv2.imread(img_dir)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

# Modelo

Definimos la instancia del modelo que vamos a usar.

```python
from cvzone.FaceDetectionModule import FaceDetector
detector = FaceDetector()
```

# Inferencia

Forward propagation del modelo y el uso de OpenCV para visualizar el mapa de características de la matriz resultante. 

```python
img, bboxs = detector.findFaces(img)

if bboxs:
    # bboxInfo - "id","bbox","score","center"
    center = bboxs[0]["center"]
    cv2.circle(img, center, 5, (255, 0, 255), cv2.FILLED)
```

# Visualización

Visualización de la imagen con el mapa de características incrustado.

```python
#Show the image with matplotlib

plt.figure(figsize=(20, 20))
plt.imshow(img)
plt.show()
```
